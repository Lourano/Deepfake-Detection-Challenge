```python
import os

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Submission = pd.read_csv('submission.csv')

Submission.hist()
```

```python
Metadata = pd.read_csv('metadata')

Metadata
```

```python
Metadata.info()
```

```python
TestVideosDirectory = "test_videos/"

TestVideos = sorted([x for x in os.listdir(TestVideosDirectory) if x[-4:] == ".mp4"])

type(TestVideos)
```

```python
TestVideoFilter = Metadata.filename.isin(TestVideos)

TestData = Metadata[TestVideoFilter]

TestData.sort_values(by = 'filename', inplace = True)

TestData.reset_index(drop = True, inplace = True)

TestData
```

```python
ZeroPointNineFilter = Submission.label >= 0.9

ZeroPointNineSubmission = Submission[ZeroPointNineFilter]

ZeroPointNineFilenames = ZeroPointNineSubmission.filename

ZeroPointNineSubmission.head(5)
```

```python
FilterMetadataForZeroPNineSubmission = TestData.filename.isin(ZeroPointNineFilenames)

MetadataForZeroPNineSubmission = TestData[FilterMetadataForZeroPNineSubmission]

MetadataForZeroPNineSubmission.split.unique()
```

```python
HashFunction = {'Filename': MetadataForZeroPNineSubmission.filename, 'Label': MetadataForZeroPNineSubmission.label}

RealTestDataOnlyForZeroPNine = pd.DataFrame(HashFunction)

RealTestDataOnlyForZeroPNine.Label.value_counts()
```

```python
SecondHashFunction = {'Filename': TestData.filename, 'RealLabel': TestData.label, 'PredictLabel': Submission.label}

ComparisonOfResults = pd.DataFrame(SecondHashFunction)

ComparisonOfResults
```

```python
FilterOfRealVideos = ComparisonOfResults.RealLabel == 'REAL'

ComparisonOfRealVideos = ComparisonOfResults[FilterOfRealVideos]

ComparisonOfRealVideos
```

```python
ComparisonOfRealVideos.hist()
```

```python
RealPredictionsFilter = (ComparisonOfRealVideos.PredictLabel <= 0.75) & (ComparisonOfRealVideos.PredictLabel >= 0.5)

RealPredictionsSet = ComparisonOfRealVideos[RealPredictionsFilter]

RealPredictionsSet
```

```python
ComparisonOfResults.RealLabel = ComparisonOfResults.RealLabel.map({'REAL': 0, 'FAKE': 1})

ComparisonOfResults
```

```python
from sklearn.metrics import log_loss

log_loss(ComparisonOfResults.RealLabel, ComparisonOfResults.PredictLabel)
```

```python
CopySet = ComparisonOfResults.copy()

log_loss(CopySet.RealLabel, CopySet.PredictLabel)
```
